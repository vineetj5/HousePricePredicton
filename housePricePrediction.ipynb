```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fPath = r'C:\Users\User\Downloads\archive\Housing.csv'
df = pd.read_csv(fPath)
df.head()
```

```python
#iloc[row,column]
#iloc[start:end , start:end]
x = df.iloc[:,:1]
y = df.iloc[:,1:5]
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg = LinearRegression()
mse = cross_val_score(lin_reg,x,y,scoring='neg_mean_squared_error',cv=10)
mean_mse = np.mean(mse)
mean_mse
```

```python
#Ridge Regression

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
# we are defining multiple value of alpha as gridsearchcv takes various combinations of alpha and gives the best fit model
params = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}

ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=10)
ridge_regressor.fit(x,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
df =load_breast_cancer()
X = pd.DataFrame(df['data'],columns=df['feature_names'])
X.head()
```

```python
# dependent feature

y=pd.DataFrame(df['target'],columns=["Target"])
y.head()
```

```python
y['Target'].value_counts()
```

balanced data set

```python
## TRAIN, TEST AND SPLIT

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
params=[{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model1 = LogisticRegression(C=100,max_iter=100)
```

```python
model=GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
