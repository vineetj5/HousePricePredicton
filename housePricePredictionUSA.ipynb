```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
filePath_Housing = r"C:\Users\User\Downloads\Housing.csv"
```

```python
housingDf = pd.read_csv(filePath_Housing)
housingDf.head()
```

```python
#DATA PRE PROCESSING 

housingDf['mainroad'] = housingDf.mainroad.eq('yes').mul(1)
housingDf['guestroom'] = housingDf.guestroom.eq('yes').mul(1)
housingDf['basement'] = housingDf.basement.eq('yes').mul(1)
housingDf['hotwaterheating'] = housingDf.hotwaterheating.eq('yes').mul(1)
housingDf['airconditioning'] = housingDf.airconditioning.eq('yes').mul(1)
housingDf['parking'] = housingDf.parking.eq('yes').mul(1)
housingDf['prefarea'] = housingDf.prefarea.eq('yes').mul(1)


housingDf
```

```python
housingDf['furnishingstatus'] = housingDf['furnishingstatus'].map( {'furnished':2 ,'semi-furnished':1,'unfurnished':0}) 
housingDf.head()
```

```python
#iloc[row,column]
#iloc[start:end , start:end]

y = housingDf.iloc[:,:1]
x = housingDf.iloc[:,1:2]
x
```

```python
housingDf.describe()
```

```python
import seaborn as sns 

plt.title('Price Distribution Plot')
sns.distplot(housingDf['price'])
plt.show()
```

```python
# Finding a relationship between price and area

plt.scatter(housingDf['area'],housingDf['price'],color='red')
plt.title('Price vs Area')
plt.xlabel('Area')
plt.ylabel('Price')

plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Price vs Area (Training Set)')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Price/Area', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score1 = r2_score(y_pred_train,y_train)
r2_score1
```

REMOVING OUTLIERS

```python
#Removing the outliers 

Q1 = housingDf['area'].quantile(0.25)
Q3 = housingDf['area'].quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out rows outside of bounds
housingDf = housingDf[(housingDf['area'] >= lower_bound) & (housingDf['area'] <= upper_bound)]
```

```python
housingDf.shape
```

```python
y = housingDf.iloc[:,:1]
x = housingDf.iloc[:,1:2]
```

```python
plt.scatter(housingDf['area'],housingDf['price'],color='red')
plt.title('Price vs Area')
plt.xlabel('Area')
plt.ylabel('Price')

plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Price vs Area (Training Set)')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Price/Area', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Price vs Area (Test Set)')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Price/Area', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score1 = r2_score(y_pred_train,y_train)
r2_score1
```

```python
mse = cross_val_score(regressor,X_train,y_train,scoring='neg_mean_absolute_error',cv=10)
mean_mse = np.mean(mse)
mean_mse
```

RIDGE AND LASSO REGRESSION

```python
from sklearn.linear_model import Ridge 
from sklearn.model_selection import GridSearchCV

ridge = Ridge()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,100]}
ridge_regressor = GridSearchCV(ridge,params,scoring = 'neg_mean_absolute_error',cv = 5)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
ridge_regressor.best_score_
```

LASSO REGRESSION

```python
from sklearn.linear_model import Lasso 
from sklearn.model_selection import GridSearchCV

lasso = Lasso()

params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,100]}
lasso_regressor = GridSearchCV(lasso,params,scoring = 'neg_mean_absolute_error',cv = 5)
lasso_regressor.fit(X_train,y_train)
print(lasso_regressor.best_params_)
lasso_regressor.best_score_
```
